```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import KMeans

```

```python
data=pd.read_csv('Countries-exercise.csv')
data
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
x=data.iloc[:,1:3]
x
```

```python
kmedia=KMeans(3)
```

```python
kmedia.fit(x)
```

```python
grupos_identificados=kmedia.fit_predict(x)
grupos_identificados
```

```python
datos_grupos=data.copy()
datos_grupos['grupos']=grupos_identificados
datos_grupos
```

```python
plt.scatter(datos_grupos['Longitude'], datos_grupos['Latitude'], c=datos_grupos['grupos'], cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
