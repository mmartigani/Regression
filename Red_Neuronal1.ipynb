```python
import numpy as np #para operaciones matematicas 
import matplotlib.pyplot as plt # graficar
from mpl_toolkits.mplot3d import Axes3D #graficar en 3d
```

```python
observations=1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
#el metodo np random uniform tiene 3 argumentos , genera nro aleatorio,
#donde cada nro tiene la misma posisbilidada de ser elegido 
zs=np.random.uniform(-10,10,(observations,1))
inputs=np.column_stack((xs,zs))
print(inputs.shape)
```

```python
noise=np.random.uniform(-1,1,(observations,1))
targets=2*xs-3*zs+5+noise
print(targets.shape)
```

```python
#graficamos 
targets=targets.reshape(observations,)
fig=plt.figure()
ax=fig.add_subplot(111)#, projection ='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
#ax.set_zlabel('Targets')
#ax.view_init(azim=100)
plt.show()
targets=targets.reshape(observations,1)
```

```python
init_range=0.1
weights=np.random.uniform(-init_range,init_range,size=(2,1))
biases=np.random.uniform(-init_range,init_range,size=1)
print(weights)
#el resultado son las 2 primeras 
print(biases)
#el resultado son la 3 fila 
```

```python
learning_rate=0.02
#tasa de aprendizaje
```

```python
for i in range(100):
    outputs=np.dot(inputs,weights)+biases#var salidas 
    deltas=outputs-targets #registra la diferencia entre la salida y los objetivos 
    #calculamos la perdida 
    loss=np.sum(deltas**2)/2/observations
    print(loss) #para saber si esta disminuyendo la perdida 
    deltas_scaled=deltas/observations
    weights=weights-learning_rate*np.dot(inputs.T,deltas_scaled)
    biases=biases -learning_rate*np.sum(deltas_scaled)
    
    #QUE HACE ESTE ALGORITMO 
    # 1-CALCULA LA SALIDA PARA LOS PESOS Y SESGOS DADOOS 
    # 2-CALCULA LA PERDIDA QUE COMPONE LA SALIDA DE LOS OBJETIVOS 
    # 3-IMPRIME LA PERDIDA PARA PODER ANALIZARLA
    # 4-ACTUALIZA LOS PESOS Y SESGOS SIGUIEN DO LA METODOLOGIA DEL DESCENDO DE LA GRADIENTE 
    
    
    
    
```
