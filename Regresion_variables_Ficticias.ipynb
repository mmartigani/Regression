```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns 
```

```python
data_ficticia=pd.read_csv('1.03. Dummies.csv')
data_ficticia
```

```python
#Attendance esta nueva variable nos indica si un estudiante asistio al 75% de las clases 
#debemos mapear estos factores de yes  y no y convertirlos en 1 y 0
data=data_ficticia.copy()
data['Attendance']=data['Attendance'].map({'Yes':1,'No':0})
data
```

```python
data.describe()
#count =cantidad 
#mean =promedio nos dice que el 46% de los alumnos asistieron 
```

```python
y=data['GPA']
x1=data[['SAT', 'Attendance']]
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
#GPA=0.6439 + 0.0014*SAT +0.2226 *DUMMIES(*0 NO ASISTIO)=GPA=0.6439 + 0.0014*SAT 
#GPA=0.6439 + 0.0014*SAT +0.2226 * DUMMES *1(SI ASISTIO)=GPA=0.8565 + 0.0014*SAT (0.8565=0.6439+0.2226)
```

```python
#GRAFICAMOS 
plt.scatter(data['SAT'],y)
yhat_no=0.6439 + 0.0014*data['SAT']#ecuaciones de cada linea 
yhat_yes=0.8565 +0.0014 *data['SAT']
fig=plt.plot(data['SAT'],yhat_no, lw=2, c='#006837')# traza la linea y su color 
fig=plt.plot(data['SAT'],yhat_yes, lw=2, c='#a50026')
plt.xlabel('SAT', fontsize=20)#coloca los nombres en cada eje 
plt.ylabel('GPA', fontsize=20)
plt.show()#muestra el grafico
```
