```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('3.12. Example.csv')
data
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x=data.copy()
```

```python
kmedias=KMeans(4)
kmedias.fit(x)
```

```python
clusters=x.copy()
clusters['clusters_pred']=kmedias.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c=clusters['clusters_pred'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
```

```python
from sklearn import preprocessing
x_scaled=preprocessing.scale(x)
x_scaled
```

```python
wscc=[]
for i in range(1,10):
    kmedias=KMeans(i)
    kmedias.fit(x_scaled)
    wscc.append(kmedias.inertia_)
wscc
    
```

```python
plt.plot(range(1,10),wscc)
plt.xlabel('numeros de grupos')
plt.ylabel('wscc')
```
