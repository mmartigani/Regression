Es una biblioteca de aprendizaje profundo desarrollada por google.

Los archivos se deben guardar con la extension .npz.

Se representan con arreglos n-dimensionales.

```python
observaciones = 1000
xs = np.random.uniform(low=-10, high=10, size=(observaciones,1))
#usamos el metodo np.random.uniform
zs = np.random.uniform(-10,10, (observaciones,1))
generated_inputs=np.column_stack((xs,zs))
noise = np.random.uniform(-1,1,(observaciones,1))
generated_targets= 2*xs-3*zs +5 +noise #funcion 
np.savez('TF_intro', inputs=generated_inputs, targets=generated_targets)
#guardamos el archivo tensorflow
```

```python
training_data=np.load('TF_intro.npz')
#cargamos el archivo tensor 
```

```python
#ahora contruimos el modelo
input_size= 2   # entradas 2
output_size= 1  # salidas 1
#tensorflow incluye keras 
#Dense = toma las entradas del modelo y calcula el producto 
#escalar de entradas y pesos y suma el sesgo
#sgd=es el metodo(descendiente del gradiente escolastico)algoritmo
#mean_squared_error=error minimo cuadrado
model=tf.keras.Sequential([
                           tf.keras.layers.Dense(output_size)
                          ])
model.compile(optimizer='sgd',loss='mean_squared_error')
model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=2)
#verboso= 0 esta en silencio 
#verboso= 1 barra de progreso
#verboso= 2 linea por epoca 
#del resultado en 100 iteraciones que serian las epochs 
#loss la perdida va cayendo osea que el algoritmo funciona
```

```python
model.layers[0].get_weights()
#model es el modelo , layers las capas que nos interesan 
#en este caso es el de la posicion 0 por el metodo get_weights()
#la salida es un tensor con 2 arreglos
#el primer array para los pesos y el 2 para los sesgos
```
